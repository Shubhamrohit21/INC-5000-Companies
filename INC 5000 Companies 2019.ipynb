```python
import pandas as pd
a=pd.read_csv("INC 5000 Companies 2019.csv")
```

```python
a.head()
```

```python
a.info()
```

```python
a.isnull().sum()
```

```python
a.dropna(inplace=True)
```

```python
a=a.reset_index()
```

```python
a.head()
```

```python
for i in range(len(a["revenue"])):
    b=a.loc[i,"revenue"]
    if "Million" in b:
        b=b.replace(" Million","")
        a.loc[i,"revenue"]=float(b)*1000000
    else:
        b=b.replace(" Billion","")
        a.loc[i,"revenue"]=float(b)*1000000000
```

```python
a.head()
```

```python
a.to_csv("INC 5000 Companies 2019.csv")
```
